```python
# Merging the national datasets for 2020-2022
all_national = pd.concat([data_national, data_2020, data_2021, data_2022])

# Convert 'Date' column to datetime format
all_national['Date'] = pd.to_datetime(all_national['Date'], format='%Y%m%d')

# Preview the structure
all_national.info(), all_national.head()
```

```python
# Using the compact national dataset
# Convert 'Date' column to datetime format
data_national['Date'] = pd.to_datetime(data_national['Date'], format='%Y-%m-%d', errors='coerce')

# Filter for Japan, USA, South Africa
countries_of_interest = ['Japan', 'United States', 'South Africa']
# Use the merged DataFrame 'all_national'
filtered_compact = all_national[all_national['CountryName'].isin(countries_of_interest)].copy() # Use .copy() to avoid SettingWithCopyWarning

# Select relevant columns
selected_columns = ['CountryName', 'Date', 'StringencyIndex', 'C1_School closing', 'H7_Vaccination policy']

# Check if all selected columns exist in the DataFrame before selecting
missing_cols = [col for col in selected_columns if col not in filtered_compact.columns]
if missing_cols:
    print(f"Warning: The following columns are missing from the DataFrame: {missing_cols}")
    # You might want to exit or handle the missing columns here.
    # For now, we will proceed with the columns that are present.
selected_columns = [col for col in selected_columns if col in filtered_compact.columns]


filtered_compact = filtered_compact[selected_columns].dropna(subset=['StringencyIndex'])

# Plot stringency index over time
plt.figure(figsize=(14, 6))
# Ensure the columns used for plotting exist
if 'Date' in filtered_compact.columns and 'StringencyIndex' in filtered_compact.columns and 'CountryName' in filtered_compact.columns:
    sns.lineplot(data=filtered_compact, x='Date', y='StringencyIndex', hue='CountryName')
    plt.title('COVID-19 Stringency Index (Merged National Dataset)')
    plt.ylabel('Stringency Index')
    plt.xlabel('Date')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("Cannot plot: Necessary columns (Date, StringencyIndex, CountryName) are not present in the filtered data.")
Use code with caution
Rate this answer

```
